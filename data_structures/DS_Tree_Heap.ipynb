```python
class MinHeap():
    def __init__(self, listinitialvalues=[]):

        self.length = 0
        self.heap = []

        for i in listinitialvalues:
            self.insert(i)

    def insert(self, val):

        self.heap.append(val)
        self.bubble_up(self.length)
        self.length += 1

    def bubble_up(self, ind):

        child_ind = ind
        parent_ind = self.parentind(ind)

        while self.heap[child_ind] < self.heap[parent_ind]:
            self.heap[child_ind], self.heap[parent_ind] = self.heap[parent_ind], self.heap[child_ind]
            child_ind, parent_ind = parent_ind, self.parentind(parent_ind)

    def bubble_down(self, ind):

        parent_ind = ind
        child_ind = self.lchildind(ind)

        while child_ind < self.length and self.heap[parent_ind] > self.heap[
                child_ind]:

            # if left child is smaller than right child point index variable to right child
            if child_ind < (self.length - 1):
                if self.heap[child_ind] > self.heap[child_ind + 1]:
                    child_ind += 1

            # if parent is smaller than child then swapping parent with child having higher value
            if self.heap[parent_ind] > self.heap[child_ind]:
                self.heap[child_ind], self.heap[parent_ind] = self.heap[
                    parent_ind], self.heap[child_ind]

            parent_ind, child_ind = child_ind, self.lchildind(child_ind)

    def delete_val(self, val):
        
        if self.heap == []:
            pass
        
        else:
            ind = self.heap.index(val)
            self.heap[ind] = self.heap.pop()
            self.length -= 1
            
            # option 1: bubble it up
            if self.heap[ind] < self.heap[self.parentind(ind)]:
                self.bubble_up(ind)
            # option 2: bubble it down
            else:
                self.bubble_down(ind)

    # utility functions
    def printmin(self):
        print(self.heap[0])

    def parentind(self, ind):
        return int((ind - 1) / 2)

    def lchildind(self, ind):
        return ind * 2 + 1

    def rchildind(self, ind):
        return ind * 2 + 2

    def parent(self, ind):
        return self.heap[self.parentind(ind)]

    def lchild(self, ind):
        return self.heap[self.lchildind(ind)]

    def rchild(self, ind):
        return self.heap[self.rchildind(ind)]

    def __str__(self):
        return ' '.join(map(str, self.heap))

    # old versions
    def bubble_up_v0(self, ind):

        parent_ind = self.parentind(ind)

        while parent_ind >= 0:

            if self.heap[parent_ind] > self.heap[ind]:
                self.heap[ind], self.heap[parent_ind] = self.heap[
                    parent_ind], self.heap[ind]
                ind = self.parentind(ind)
                parent_ind = self.parentind(ind)
            else:
                break

    def bubble_down_v0(self, ind):

        val = self.heap[ind]

        if self.lchildind(ind) > (self.length - 1):
            return (ind, True)

        elif self.rchildind(ind) < self.length:

            valh1 = self.lchild(ind)
            valh2 = self.rchild(ind)

            valreplace = min(val, valh1, valh2)

            if valreplace == val:
                return (ind, True)

            elif valreplace == valh1:
                self.heap[ind] = valh1
                self.heap[self.lchildind(ind)] = val
                return (ind, False)

            else:
                self.heap[ind] = valh2
                self.heap[self.rchildind(ind)] = val
                return (ind, False)

        elif self.lchildind(ind) < self.length:

            valh1 = self.heap[self.lchildind(ind)]
            valreplace = min(val, valh1)

            if valreplace == val:
                return (ind, True)

            elif valreplace == valh1:
                self.heap[ind] = valh1
                self.heap[2 * ind + 1] = val
                return (ind, False)
    
    def delete_val_v0(self, val):

        if self.heap == []:
            self.length = 0
            return []

        i = 0
        while i < self.length:

            if self.heap[i] == val:

                # caso 0: que el ultimo elemento sea que el hay que eliminar
                if i == self.length - 1:
                    self.heap.pop()
                    break

                # eliminar el Ãºltimo elemento
                self.heap[i] = self.heap.pop()
                self.length -= 1
                ind = i

                # caso 1: llevarlo hacia arriba (bubble it up)
                if self.parent(ind) > self.heap[ind] and self.parentind(
                        ind) >= 0:
                    self.heap = self.bubble_up(ind)

                # caso 2: llevarlo hacia abajo (reemplazandolo por el menor hijo)
                else:
                    self.bubble_down(ind)
                    #Isfinished = False
                    #while Isfinished == False:
                    #    (ind, Isfinished) = self.bubble_down(ind)
            i += 1

heap1 = MinHeap()
heap1.insert(3)
heap1.insert(4)
heap1.insert(5)
heap1.insert(8)
heap1.insert(2)
heap1.insert(1)
heap1.insert(9)
heap1.insert(7)
heap1.insert(6)
print(heap1)
heap1.delete_val(1)
heap1.delete_val(3)
print(heap1)
heap1.heap[0] = 12
heap1.bubble_down(0)
print(heap1)
```
