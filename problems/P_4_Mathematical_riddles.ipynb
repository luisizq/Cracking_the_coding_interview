# Different bits:

https://practice.geeksforgeeks.org/problems/find-sum-of-different-corresponding-bits-for-all-pairs/0 

```python
def different_bits(a, b):
    
    counter = 0
    
    remainder_a, remainder_b = 0, 0
    quotient_a, quotient_b = a, b
    
    while quotient_a or quotient_b:
        #remainder_a, quotient_a = quotient_a%2, quotient_a//2
        #remainder_b, quotient_b = quotient_b%2, quotient_b//2
        quotient_a, remainder_a = divmod(quotient_a, 2)
        quotient_b, remainder_b = divmod(quotient_b, 2)
        if remainder_a != remainder_b:
            counter += 1
    return counter

def obtain_sum(alist):
    s = 0
    for i in range(len(alist)):
        for j in range(i+1,len(alist)):
            s += 2*different_bits(alist[i], alist[j])
    return s

a = [1, 3, 5] 
print(obtain_sum(a))
```

# Minimum cut squares
- https://practice.geeksforgeeks.org/problems/min-cut-square/0

```python
from collections import defaultdict

def divisors(a):
    div = []
    for i in range(1,int(a**0.5)+1):
        if a%i == 0:
            div.append(i)
            div.append(a//i)
    return div

# solution to the problem
def find_minimum_helper(a, b, memory):
    
    if a == 0 or b == 0:
        return 0
    
    low, high = min(a, b), max(a, b)
    
    ilow, ihigh = low - 1, high - 1
    
    div = divisors(low)    
    values = []
    
    for i in div:
        
        if high - i == 0:
            values.append(low//i)
            
        elif not memory[ilow][ihigh - i]:
            memory[ilow][ihigh - i] = find_minimum_helper(low, high-i, memory)     
            values.append(low//i + memory[ilow][ihigh-i])

        else:
            values.append(low//i + memory[ilow][ihigh-i])

    
    return min(values)
    
    

def find_minimum(a, b):
    
    a, b = max(a, b), min(a, b)
    
    memory = [[None]*a for _ in range(b)]
    
    for i in range(a):
        memory[0][i] = i + 1
    for i in range(b):
        memory[i][0] = i + 1

    return find_minimum_helper(a, b, memory)

print(find_minimum(30,36))
```

# Word boogle 

```python
def neighbours(i, j, n, m):
    
    for auxi in [-1,0,1]:
        for auxj in [-1,0,1]:
            if 0 <= auxi + i < n and 0 <= auxj + j < m and (auxi != 0 or auxj != 0):
                yield (auxi + i, auxj + j)
    
    
def find_word(i, j, boogle, word, coordinates_visited):
    
    if len(word) == 0:
        return True
    
    n = len(boogle)
    m = len(boogle[0])
    
    
    for a, b in neighbours(i, j, n, m):
        if (a, b) not in coordinates_visited:
            #print(word, (a, b), coordinates_visited)
            if boogle[a][b] == word[0]:
                is_found = find_word(a, b, boogle, word[1:], coordinates_visited+[(a,b)])
                if is_found:
                    return True
    return False
    

def find_list_words(i, j, boogle, dictionary):
    
    list_of_words = []
    
    initial_letter = boogle[i][j]
    
    for word in dictionary:
        if word[0] == initial_letter:
                
            coordinates_visited = [(i,j)]
            is_found = find_word(i, j, boogle, word[1:], coordinates_visited)
            
            if is_found:
                list_of_words += [word]

    return list_of_words
    

def word_boggle(dictionary, boogle):
    
    dictionary_aux = dictionary.copy()
    words_found = []
    
    n = len(boogle)
    m = len(boogle[0])
    
    for i in range(n):
        for j in range(m):
            
            words = find_list_words(i, j, boogle, dictionary_aux)
            
            for word in words:
                dictionary_aux.discard(word)

            words_found += words
            
    return words_found
            
```

```python
dictionary = {"GEEKS", "FEEL", "FEELASUG", "QUIZ", "GO"}
boggle = [['G','E','L'],
          ['U','E','A'],
          ['F','S','K']]

word_boggle(dictionary, boggle)
```
